```python
## Importing important libraries:

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn import preprocessing


from sklearn import metrics
from sklearn.metrics import accuracy_score as acc


from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict


from sklearn.metrics import make_scorer, accuracy_score
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
## Dropping unnecessary columns from the dataset:

data.drop(columns=['Unnamed: 0','match_event_id','game_season','team_name','date_of_game','shot_id_number','lat/lng','type_of_shot','type_of_combined_shot','match_id','team_id',],inplace=True)
data=data.iloc[:,:-5]
```

```python
## Filling the reamining missing values:

data['location_x'].fillna(data['location_x'].mean(), inplace=True)
data['location_y'].fillna(data['location_y'].mean(), inplace=True)
data['remaining_min'].fillna(data['remaining_min'].mean(), inplace=True)
data['power_of_shot'].fillna(data['power_of_shot'].median(), inplace=True)
data['knockout_match'].fillna(data['knockout_match'].median(), inplace=True)
data['remaining_sec'].fillna(data['remaining_sec'].mean(), inplace=True)
data['distance_of_shot'].fillna(data['distance_of_shot'].mean(), inplace=True)
data['shot_basics'].fillna(value='Mid Range', inplace=True)
data['range_of_shot'].fillna(value='Less Than 8 ft.', inplace=True)
data['area_of_shot'].fillna(value='Center(C)', inplace=True)
data['home/away'].fillna(value='MANU @ SAS', inplace=True)
```

```python
data.isnull().sum()
```

```python
## Transforming the feature 'home/away' into a categorical variable having values only 'home' and 'away' :

for i in range(data.shape[0]):
    if('@' in data.loc[i,'home/away']):
        data.loc[i,'home/away']='away'
        
    if('vs' in data.loc[i,'home/away']):
        data.loc[i,'home/away']='home'
    
   
```

```python
data.head()
```

```python
## Splitting the data into train and test dataset:

test=data[data['is_goal'].isnull()]
train=data[data['is_goal'].notnull()]
```

```python
data.dtypes
```

```python
data.columns
```

```python
### PREPARING y, X AND X_test:

y=train['is_goal'].values

X=train[['location_x', 'location_y', 'remaining_min', 'power_of_shot','knockout_match', 'remaining_sec', 'distance_of_shot','area_of_shot', 'shot_basics', 'range_of_shot', 'home/away']].values
X_test=test[['location_x', 'location_y', 'remaining_min', 'power_of_shot','knockout_match', 'remaining_sec', 'distance_of_shot','area_of_shot', 'shot_basics', 'range_of_shot', 'home/away']].values
```

```python
### LABEL ENCODING:

area_of_shot=preprocessing.LabelEncoder()
shot_basics=preprocessing.LabelEncoder()
range_of_shot=preprocessing.LabelEncoder()
last=preprocessing.LabelEncoder()

## FOR Area_of_shot:

area_of_shot.fit(['Mid Ground(MG)','Right Side(R)', 'Left Side(L)', 'Left Side Center(LC)','Right Side Center(RC)', 'Center(C)'])

X[:,-4]=area_of_shot.transform(X[:,-4])
X_test[:,-4]=area_of_shot.transform(X_test[:,-4])


## FOR Shot_basics:

shot_basics.fit(['Mid Range','Mid Ground Line', 'Right Corner', 'Left Corner' ,'Goal Area', 'Goal Line', 'Penalty Spot'])

X[:,-3]=shot_basics.transform(X[:,-3])
X_test[:,-3]=shot_basics.transform(X_test[:,-3])


## FOR Range_of_shot:

range_of_shot.fit(['Back Court Shot','24+ ft.','16-24 ft.', '8-16 ft.', 'Less Than 8 ft.'])

X[:,-2]=range_of_shot.transform(X[:,-2])
X_test[:,-2]=range_of_shot.transform(X_test[:,-2])

## FOR last column:

last.fit(['away','home'])

X[:,-1]=last.transform(X[:,-1])
X_test[:,-1]=last.transform(X_test[:,-1])
```

```python
### STANDARDISING X AND X_TEST:


X_=preprocessing.StandardScaler().fit(X).transform(X)
X_test_=preprocessing.StandardScaler().fit(X_test).transform(X_test)
```

```python
### TEST TRAIN SPLIT:

x_train,x_test_,Y_train,Y_test=train_test_split(X_,y,test_size=0.3,random_state=0) 
```

```python
### LOGISTIC REGRESSION:

from sklearn.linear_model import LogisticRegression

lr=LogisticRegression(C=5, solver='liblinear').fit(x_train,Y_train)

yhat=lr.predict(x_test_)

acc(Y_test,yhat)*100
```

```python
# KNN:

from sklearn.neighbors import KNeighborsClassifier

k=1000


neigh=KNeighborsClassifier(n_neighbors=k).fit(x_train,Y_train)

yhat1=neigh.predict(x_test_)

acc(Y_test,yhat1)*100
```

```python
### SVM (RBF)

from sklearn import svm

clf3=svm.SVC(kernel='rbf',gamma=0.05, C=8, probability=True)


clf3.fit(x_train,Y_train)

yhat3=clf3.predict(x_test_)

(acc(Y_test,yhat3))*100
```

```python
### DECISION TREE:

from sklearn.tree import DecisionTreeClassifier

tree=DecisionTreeClassifier(criterion='entropy', max_depth=4)



tree.fit(x_train,Y_train)

yhat2=tree.predict(x_test_)

acc(Y_test,yhat2)*100
```

```python
### RANDOM FOREST CLASSIFIER

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import make_scorer, accuracy_score
from sklearn.model_selection import GridSearchCV

# Choose the type of classifier. 
rnd = RandomForestClassifier()


### TO GET PARAMETERS OF ANY MODEL:

rnd.get_params().keys()

```

```python
### GRIDSEARCH + RANDOM FOREST CLASSIFIER:


# Choose some parameter combinations to try
parameters = {'n_estimators': [4, 6, 9], 
              'max_features': ['log2', 'sqrt','auto'], 
              'criterion': ['entropy', 'gini'],
              'max_depth': [2, 3,5, 6, 10], 
              'min_samples_split': [2, 3, 5],
              'min_samples_leaf': [1,5,8]
             }

# Type of scoring used to compare parameter combinations
acc_scorer = make_scorer(accuracy_score)

# Run the grid search
grid_obj = GridSearchCV(rnd, parameters, scoring=acc_scorer)
grid_obj = grid_obj.fit(x_train, Y_train)

# Set the clf to the best combination of parameters
rnd = grid_obj.best_estimator_  

# Fit the best algorithm to the data. 
rnd.fit(x_train, Y_train)

yhat7=rnd.predict(x_test_)

(acc(Y_test,yhat7))*100
```

```python
### VOTING CLASSIFIER(ENSEMBLE)

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.svm import SVC



voting_clf=VotingClassifier(estimators=[('lr',lr),('rf', rnd),('dt',tree),('knn',neigh),('rbf',clf3)], voting='soft')

voting_clf.fit(x_train,Y_train)

yhat6=voting_clf.predict(x_test_)

(acc(Y_test,yhat6))*100
```
